```python
# Sigmoid function:

def sigmoid(z):
    return 1/(1 + np.exp(-z)) 

def dense_layer(a_in,W,b):
    units = W.shape[1]
    a_out = np.zeros(units)
    for j in range(units):
        w = W[:,j] # first column
        z = np.dot(w,a_in) + b[j]
        a_out[j] = sigmoid(z)
    
    return a_out
```

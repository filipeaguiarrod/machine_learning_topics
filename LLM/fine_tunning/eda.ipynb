```python
import pandas as pd
```

```python
df = pd.read_csv('credit_card_transactions_202503111548.csv')
```

```python
def padronizar_texto(texto):
  
  texto = str(texto)
  """Padroniza o texto removendo espaços extras e capitalizando a primeira letra de cada palavra."""
  texto = ' '.join(texto.split())  # Remove espaços extras
  texto = texto.title()  # Capitaliza a primeira letra de cada palavra
  return texto
```

```python
df['categoria'] = df['categoria'].apply(lambda x: padronizar_texto(x))
```

```python
df['categoria'].value_counts(normalize=True)
```

```python
grouped = df.groupby('categoria')
```

```python
grouped
```

```python
sampled_df = grouped.apply(lambda x: x.sample(n=min(len(x), 25)))
```

```python
sampled_df['categoria'].value_counts()
```
